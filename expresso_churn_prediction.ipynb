```python
## filling missing values in the categorical collumns using the git guy method
train['TOP_PACK'] = train['TOP_PACK'].fillna(value= 'none')
train['REGION'] = train['REGION'].fillna(value= 'none')

test['TOP_PACK'] = test['TOP_PACK'].fillna(value= 'none')
test['REGION'] = test['REGION'].fillna(value= 'none')
```

```python
## filling the numerical values
for i in ['MONTANT', 'FREQUENCE_RECH', 'REVENUE', 'ARPU_SEGMENT', 'FREQUENCE', 'DATA_VOLUME', 'ON_NET', 'ORANGE', 'TIGO', 'ZONE1', 'ZONE2', 'TOP_PACK', 'FREQ_TOP_PACK' ]:
    train[i] = train[i].fillna(value = -999999)
    test[i] = test[i].fillna(value = -9999999)
```

```python
X  = train.drop(['user_id', 'MRG', 'CHURN',], axis=1)
y = train['CHURN']

X_test = test.drop(['user_id', 'MRG',], axis = 1)
```
